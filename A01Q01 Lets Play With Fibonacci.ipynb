Question 1

Write a Python program to get the Fibonacci series between 0 to 50

```python
print("Fib Series Between 0 to 50 Is : ")
Number1 = 0
Number2 = 1
print(Number1)
print(Number2)
for i in range(0,51):
    Number3=Number1+Number2
    Number1=Number2
    Number2=Number3
    if Number3<50:
        print(Number3)
```
